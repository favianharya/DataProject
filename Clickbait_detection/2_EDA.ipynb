```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import wordcloud
```

```python
df = pd.read_csv("clickbait_data_clean.csv")
```

```python
label_counts = df['clickbait'].value_counts()

# Replace label values with respective names
label_counts.index = label_counts.index.map({1: 'Clickbait', 0: 'Non-Clickbait'})

sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.25)

plt.figure(figsize=(8, 8))
pie_chart = plt.pie(label_counts, labels=label_counts.index, autopct=lambda p: '{:.1f}%\n({:d})'.format(p, int(round(p * sum(label_counts) / 100))),
                    startangle=90, colors=sns.color_palette('pastel'))

plt.title('Label Pie Chart', fontsize=16)
plt.show()
```

```python
df_clickbait = df[df['clickbait'] == 1].reset_index()
df_nonclickbait = df[df['clickbait'] == 0].reset_index()
```

```python
clickbait_word = {}
for i in range(len(df_clickbait)):
  temp_text = df_clickbait.loc[i]['clean_text']
  temp_split = temp_text.split(' ')
  temp_set = list(set(temp_split))
  for j in range(len(temp_set)):
    try:
      clickbait_word[temp_set[j]] += 1
    except:
      clickbait_word[temp_set[j]] = 1

clickbait_word_sorted = dict(sorted(clickbait_word.items(), key=lambda x: x[1], reverse=True))
clickbait_word_sorted
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='white')

wordcloud.generate_from_frequencies(clickbait_word_sorted)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
non_clickbait_word = {}
for i in range(len(df_nonclickbait)):
  temp_text = df_nonclickbait.loc[i]['clean_text']
  temp_split = temp_text.split(' ')
  temp_set = list(set(temp_split))
  for j in range(len(temp_set)):
    try:
      non_clickbait_word[temp_set[j]] += 1
    except:
      non_clickbait_word[temp_set[j]] = 1

non_clickbait_word_sorted = dict(sorted(non_clickbait_word.items(), key=lambda x: x[1], reverse=True))
non_clickbait_word_sorted
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='white')

wordcloud.generate_from_frequencies(non_clickbait_word_sorted)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Add a check for missing values before calculating the length
df['text_length'] = df['clean_text'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)

# Find the text length with the maximum frequency
max_length = df['text_length'].value_counts().idxmax()

# Calculate the maximum frequency
max_frequency = df['text_length'].value_counts().max()

# Display the maximum text length and its frequency
print("Text Length with Maximum Frequency:", max_length)
print("Maximum Frequency:", max_frequency)

# Plot the distribution of text length
plt.figure(figsize=(10, 6))
sns.histplot(df['text_length'], bins=50, kde=True)
plt.title('Distribution of Text Length')
plt.xlabel('Text Length')
plt.ylabel('Frequency')

# Annotate the plot with the maximum text length and its frequency
plt.annotate(f'Max Length: {max_length}\nFrequency: {max_frequency}',
             xy=(max_length, max_frequency),
             xytext=(max_length + 10, max_frequency - 50),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=13, color='red')

plt.show()
```
